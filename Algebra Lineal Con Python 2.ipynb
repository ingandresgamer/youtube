```python
import numpy as np
import matplotlib.pyplot as plt 
```

```python
plt.quiver([1],[1],[2],[0.5],color='blue',angles='xy',scale_units='xy',scale=1)
plt.quiver([0],[0],[2],[0.5],color='red',angles='xy',scale_units='xy',scale=1)
plt.xlim(-1,5)
plt.ylim(-1,5)
plt.axvline(x=0,color='grey')
plt.axhline(y=0,color='grey')
```

```python
def plotvector(vecs,colors):
    plt.figure()
    plt.axvline(x=0,color='grey')
    plt.axhline(y=0,color='grey')
    for i in range(len(vecs)):
        x=np.concatenate([[0,0],vecs[i]])
        plt.quiver(x[0],x[1],x[2],x[3],color=colors[i],angles='xy',scale_units='xy',scale=1)
```

```python
v1=np.array([2,2])
v2=np.array([-2,2])

plotvector([v1,v2],colors=['red','blue'])
plt.xlim(-3,3)
plt.ylim(-3,3)            
plt.grid()
```

```python
A=np.array([[2,4],[-4,2]])
print(A)

v3=A.dot(v1)
print(v3)
plotvector([v1,v2,v3],colors=['red','blue','green'])
plt.xlim(-14,14)
plt.ylim(-14,14)            
plt.grid()
```
