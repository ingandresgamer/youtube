```python
import numpy as np 
import matplotlib.pyplot as plt
```

```python
def plotvector(vecs,colors):
    plt.figure()
    plt.axvline(x=0,color='grey')
    plt.axhline(y=0,color='grey')
    for i in range(len(vecs)):
        x=np.concatenate([[0,0],vecs[i]])
        plt.quiver(x[0],x[1],x[2],x[3],color=colors[i],angles='xy',scale_units='xy',scale=1)
```

```python
A=np.array([[3,2],[4,1]])
print(A)
eigv,eigvec=np.linalg.eig(A)
print(f'Los Autovalores de A so: {eigv}')
print(f'El autovector asociado al autovalor {eigv[0]}  es {eigvec[:,0]}')
print(f'El autovector asociado al autovalor {eigv[1]}  es {eigvec[:,1]}')
```

```python
v=np.array([[1],[-1]])
v1=eigvec[:,0]
```

```python
B=A.dot(v)
C=A.dot(v1)
print(B)
print(C)
plotvector([v.flatten(),C.flatten(),B.flatten(),v1.flatten()],colors=['green','orange','red','blue'])
plt.xlim(-4,4)
plt.ylim(-4,4)
plt.xlabel('X1')
plt.ylabel('X2')
plt.grid()
```
